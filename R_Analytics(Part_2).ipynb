```r
# ============================================================
# CELL 3: DATA CLEANING — Fix inconsistent zone names
# This is IMPORTANT — without this, GROUP BY zone gives wrong results
# ============================================================
clean_zone <- function(z) {
  z <- toupper(trimws(z))
  z <- ifelse(z == "CTR", "CENTRAL", z)
  z <- ifelse(z == "RIVERSIDE", "RIVERSIDE", z)
  return(z)
}

customers$home_zone   <- clean_zone(customers$home_zone)
drivers$base_zone     <- clean_zone(drivers$base_zone)
vehicles$assigned_zone <- clean_zone(vehicles$assigned_zone)
orders$pickup_zone    <- clean_zone(orders$pickup_zone)
orders$dropoff_zone   <- clean_zone(orders$dropoff_zone)
hubs$zone             <- clean_zone(hubs$zone)

cat("Zone cleaning complete. Unique zones in customers:\n")
print(table(customers$home_zone))
```

```r
# Count each delivery status
status_counts <- as.data.frame(table(deliveries$delivery_status))
colnames(status_counts) <- c("Status", "Count")

ggplot(status_counts, aes(x = Status, y = Count, fill = Status)) +
  geom_bar(stat = "identity") +
  scale_fill_manual(values = c("OnTime" = "steelblue",
                                "Delayed" = "orange",
                                "Failed" = "red")) +
  labs(
    title = "NorthStar Delivery Outcomes",
    x = "Delivery Status",
    y = "Number of Deliveries"
  ) +
  theme_minimal()

# WRITE IN YOUR REPORT:
# Out of 950 deliveries, 132 failed (13.9%) and 202 were delayed (21.3%).
# This means only 64.8% of deliveries were completed on time.
```

```r
inc_counts <- as.data.frame(table(incidents$incident_type))
colnames(inc_counts) <- c("Type", "Count")
inc_counts <- inc_counts[order(-inc_counts$Count), ]

ggplot(inc_counts, aes(x = reorder(Type, Count), y = Count, fill = Type)) +
  geom_bar(stat = "identity") +
  geom_text(aes(label = Count), hjust = -0.2) +
  coord_flip() +
  labs(title = "NorthStar: Incident Types",
       subtitle = "Source: incidents.csv (280 records)",
       x = "Incident Type", y = "Count") +
  theme_minimal() +
  theme(legend.position = "none")

# WRITE IN REPORT:
# ProofMissing (46) and CustomerNoShow (44) are the most common incidents.
# BatteryAlert (36) and VehicleFault (37) together = 73 vehicle-related incidents.
# This shows poor vehicle maintenance is a real operational problem.
```

```r
# Remove missing battery health values
veh_clean <- vehicles[!is.na(vehicles$battery_health_pct), ]

ggplot(veh_clean, aes(x = maintenance_status,
                       y = battery_health_pct,
                       fill = maintenance_status)) +
  geom_boxplot() +
  scale_fill_manual(values = c("Active" = "#2ecc71",
                                "InRepair"  = "#e74c3c",
                                "Scheduled" = "#f39c12")) +
  labs(title = "Vehicle Battery Health by Maintenance Status",
       subtitle = "Source: vehicles.csv (120 records)",
       x = "Maintenance Status", y = "Battery Health (%)") +
  theme_minimal() +
  theme(legend.position = "none")

# WRITE IN YOUR REPORT:
# Vehicles with InRepair status have lower battery health on average.
# This confirms that battery degradation is a key reason vehicles go InRepair.
# NorthStar needs to monitor battery health proactively before vehicles break down.
```

```r
# Calculate failure rate per event type
app_summary <- app_events %>%
  group_by(event_type) %>%
  summarise(
    total = n(),
    failed = sum(success_flag == 0),
    avg_latency = round(mean(api_latency_ms), 0)
  ) %>%
  mutate(fail_pct = round(100 * failed / total, 1))

print(app_summary)

ggplot(app_summary, aes(x = reorder(event_type, -fail_pct),
                         y = fail_pct, fill = event_type)) +
  geom_bar(stat = "identity") +
  geom_text(aes(label = paste0(fail_pct, "%")), vjust = -0.4) +
  labs(title = "App Event Failure Rate by Event Type",
       subtitle = "Source: app_events.csv (640 records)",
       x = "Event Type", y = "Failure Rate (%)") +
  theme_minimal() +
  theme(legend.position = "none",
        axis.text.x = element_text(angle = 25, hjust = 1))

# WRITE IN YOUR REPORT:
# Some event types fail more often than others.
# High failure rates on booking or tracking events directly frustrate customers
# before their order even begins.
```

```r
# Join complaints → orders → to get pickup_zone
comp_zone <- merge(complaints,
                   orders[, c("order_id", "pickup_zone")],
                   by = "order_id", all.x = TRUE)

comp_zone$pickup_zone <- toupper(trimws(comp_zone$pickup_zone))
comp_zone$pickup_zone[comp_zone$pickup_zone == "CTR"] <- "CENTRAL"
comp_zone <- comp_zone[!is.na(comp_zone$pickup_zone), ]

zone_comp <- as.data.frame(table(comp_zone$pickup_zone))
colnames(zone_comp) <- c("Zone", "Complaints")

ggplot(zone_comp, aes(x = reorder(Zone, -Complaints),
                       y = Complaints, fill = Zone)) +
  geom_bar(stat = "identity") +
  geom_text(aes(label = Complaints), vjust = -0.4) +
  labs(title = "Number of Complaints by Zone",
       subtitle = "Source: complaints.csv + orders.csv",
       x = "Pickup Zone", y = "Number of Complaints") +
  theme_minimal() +
  theme(legend.position = "none",
        axis.text.x = element_text(angle = 30, hjust = 1))
```

```r
# Join hubs with deliveries to see if capacity score affects delivery count
hub_del <- merge(hubs,
                 aggregate(delivery_id ~ hub_id, data = deliveries, FUN = length),
                 by = "hub_id", all.x = TRUE)
colnames(hub_del)[ncol(hub_del)] <- "total_deliveries"
hub_del$total_deliveries[is.na(hub_del$total_deliveries)] <- 0

ggplot(hub_del, aes(x = capacity_score, y = total_deliveries,
                     label = hub_name, colour = hub_type)) +
  geom_point(size = 4) +
  geom_text(vjust = -0.8, size = 3) +
  labs(title = "Hub Capacity Score vs Total Deliveries Handled",
       subtitle = "Source: hubs.csv + deliveries.csv",
       x = "Hub Capacity Score", y = "Total Deliveries",
       colour = "Hub Type") +
  theme_minimal()

# WRITE IN YOUR REPORT:
# Are high-capacity hubs actually handling more deliveries?
# If a hub has a high capacity score but low deliveries, it is underused.
# If a low-capacity hub has very high deliveries, it may be overwhelmed.
```
